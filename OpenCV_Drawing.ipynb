```python
import cv2
```

```python
im = input("Enter the path of image: ")
original = cv2.imread(im)
```

```python
if original is not None:

    while True:
        image = original.copy()   # refresh image every option
        print("\nAll Options:")
        print("1: Line\n2: Rectangle\n3: Circle\n4: Text\n5: Exit")

        options = int(input("Select option: "))

        if options == 5:
            print("Successfully logged out!")
            break

        elif options == 1:
            x1, y1 = map(int, input("Enter x y of point1: ").split())
            x2, y2 = map(int, input("Enter x y of point2: ").split())
            color = tuple(map(int, input("Enter RGB: ").split()))
            thickness = int(input("Enter thickness: "))
            cv2.line(image, (x1, y1), (x2, y2), color, thickness)

        elif options == 2:
            x1, y1 = map(int, input("Enter x y of point1: ").split())
            x2, y2 = map(int, input("Enter x y of point2: ").split())
            color = tuple(map(int, input("Enter RGB: ").split()))
            thickness = int(input("Enter thickness: "))
            cv2.rectangle(image, (x1, y1), (x2, y2), color, thickness)

        elif options == 3:
            x, y = map(int, input("Enter center x y: ").split())
            radius = int(input("Enter radius: "))
            color = tuple(map(int, input("Enter RGB: ").split()))
            thickness = int(input("Enter thickness: "))
            cv2.circle(image, (x, y), radius, color, thickness)

        elif options == 4:
            text = input("Enter text: ")
            x, y = map(int, input("Enter x y for text: ").split())
            font_scale = float(input("Enter the size of font (1.0 is normal & 2.0 is big):")) #1.0 is normal and 2.0 is big
            color = tuple(map(int, input("Enter RGB: ").split()))
            thickness = int(input("Enter thickness: "))
            cv2.putText(
                image,
                text,
                (x, y),
                cv2.FONT_HERSHEY_SIMPLEX,
                font_scale,
                color,
                thickness
            )

        else:
            print("Invalid option")
            continue

        # SHOW IMAGE
        cv2.imshow("Drawing Window", image)

        # WAIT & CLOSE WINDOW
        cv2.waitKey(0)
        cv2.destroyWindow("Drawing Window")

        # SAVE OPTION
        save = input("Do you want to save image? (y/n): ").lower()
        if save == "y":
            name = input("Enter image name with extension: ")
            cv2.imwrite(name, image)
            print("Image saved successfully!")

    cv2.destroyAllWindows()

else:
    print("Image not found!")
```
